```python
from ml4a import image
from ml4a.models import deepdream

channels = deepdream.get_random_favorites(deepdream.layers_m3, 1)

config = {
    'objective': channels,
    'num_octaves': 3,
    'octave_ratio': 1.333,
    'num_iterations': 30,
    'size': 512
}

input_img = image.monalisa()
output_img = deepdream.run(config, input_img)
image.display(image.concatenate_images([input_img, output_img]))
```
